```python
# Завдання 1. Створіть одновимірний масив (вектор) з перших 10-ти додатних парних чисел та виведіть його значення.

import numpy as np

vector = np.arange(2, 21, 2)

print("Вектор з перших 10 додатних парних чисел:")
print(vector)
```

```python
# Завдання 2. Створіть двовимірний масив (матрицю) розміром 3x3,
# заповніть його сумою кожного числа та його попереднього, починаючи з 1, та виведіть його значення.

import numpy as np

values = []
for i in range(1, 10):
    values.append(i + (i - 1))

matrix = np.array(values).reshape(3, 3)

print("Матриця 3x3:")
print(matrix)
```

```python
# Завдання 3. Створіть масив розміром 2x3, заповніть його випадковими цілими числами
# в діапазоні від -1 до 1 та виведіть його значення.

import numpy as np

matrix = np.random.randint(-1, 2, size=(2, 3))

print("Матриця 2x3 зі значеннями від -1 до 1:")
print(matrix)
```

```python
# Завдання 4. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами
# в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

import numpy as np

matrix = np.random.rand(5, 5)

row_sums = np.sum(matrix, axis=1)

column_vector = row_sums.reshape(5, 1)

print("Матриця 5x5:")
print(matrix)

print("\nВектор-стовпчик (сума елементів рядків):")
print(column_vector)
```

```python
# Завдання 5. Створіть два одновимірних масиви розміром 5,
# заповніть їх випадковими дійсними числами від 0 до 1
# та виконайте на них поелементні операції додавання, віднімання та множення.

import numpy as np

a = np.random.rand(5)
b = np.random.rand(5)

add = a + b
sub = a - b
mul = a * b

print("Масив A:", a)
print("Масив B:", b)

print("\nПоелементне додавання (A + B):", add)
print("Поелементне віднімання (A - B):", sub)
print("Поелементне множення (A * B):", mul)
```

```python
# Завдання 6. Створіть дві матриці 3x3, заповніть їх довільними дійсними числами від 5 до 30
# та знайдіть їх добуток.

import numpy as np

A = np.random.uniform(5, 30, (3, 3))
B = np.random.uniform(5, 30, (3, 3))

C = np.dot(A, B)

print("Матриця A:\n", A)
print("\nМатриця B:\n", B)
print("\nДобуток матриць A * B:\n", C)
```

```python
# Завдання 7. Створіть дві матриці розміром 2x3 та 3x4,
# заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.

import numpy as np

A = np.random.randint(1, 6, size=(2, 3))  # 2x3
B = np.random.randint(1, 6, size=(3, 4))  # 3x4

C = np.dot(A, B)

print("Матриця A (2x3):\n", A)
print("\nМатриця B (3x4):\n", B)
print("\nДобуток матриць A * B (2x4):\n", C)
```

```python
# Завдання 8. Створіть матрицю розміром 3x3,
# заповніть її випадковими цілими числами в діапазоні від 1 до 8
# та знайдіть її обернену матрицю.

import numpy as np

A = np.random.randint(1, 9, size=(3, 3))

print("Матриця A:")
print(A)

det = np.linalg.det(A)

if det != 0:
    A_inv = np.linalg.inv(A)
    print("\nОбернена матриця A⁻¹:")
    print(A_inv)
else:
    print("\nМатриця є виродженою (det=0), оберненої матриці не існує.")
```

```python
# Завдання 9. Створіть матрицю розміром 4x4,
# заповніть її випадковими дійсними числами в діапазоні від 0 до 1
# та транспонуйте її.

import numpy as np

A = np.random.rand(4, 4)

print("Матриця A:")
print(A)

A_T = A.T

print("\nТранспонована матриця Aᵀ:")
print(A_T)
```

```python
# Завдання 11. Створіть матрицю розміром 2x3 та вектор розміром 3x1,
# заповніть матрицю випадковими цілими числами (1..10),
# вектор випадковими дійсними числами (0..1),
# знайдіть добуток матриці на вектор.

import numpy as np

A = np.random.randint(1, 11, size=(2, 3))

v = np.random.rand(3, 1)

result = np.dot(A, v)

print("Матриця A (2x3):\n", A)
print("\nВектор v (3x1):\n", v)
print("\nДобуток A * v (2x1):\n", result)
```

```python
# Завдання 12. Створіть дві матриці розміром 7x7,
# заповніть їх випадковими цілими числами в діапазоні від 1 до 10
# та знайдіть їхню різницю.

import numpy as np

A = np.random.randint(1, 11, size=(7, 7))
B = np.random.randint(1, 11, size=(7, 7))

C = A - B

print("Матриця A:\n", A)
print("\nМатриця B:\n", B)
print("\nРізниця матриць (A - B):\n", C)
```

```python
# Завдання 13. Створіть дві матриці розміром 7x7,
# заповніть їх випадковими цілими числами в діапазоні від -10 до 10
# та знайдіть суму їх елементів.

import numpy as np

A = np.random.randint(-10, 11, size=(7, 7))
B = np.random.randint(-10, 11, size=(7, 7))

sum_A = np.sum(A)
sum_B = np.sum(B)

total_sum = sum_A + sum_B

print("Матриця A:\n", A)
print("\nСума елементів матриці A:", sum_A)

print("\nМатриця B:\n", B)
print("\nСума елементів матриці B:", sum_B)

print("\nЗагальна сума елементів обох матриць:", total_sum)
```

```python
# Завдання 14. Створіть матрицю розміром 3x3,
# заповніть її випадковими цілими числами в діапазоні від 1 до 10
# та знайдіть середнє арифметичне значення елементів матриці.

import numpy as np

A = np.random.randint(1, 11, size=(3, 3))

mean_value = np.mean(A)

print("Матриця A:\n", A)
print("\nСереднє арифметичне значення елементів:", mean_value)
```

```python
# Завдання 15. Створіть матрицю розміром 8x8,
# заповніть її випадковими дійсними числами в діапазоні від -1 до 1
# та знайдіть вектор-стовпчик, що містить найменший елемент у кожному рядку.

import numpy as np

A = np.random.uniform(-1, 1, size=(8, 8))

min_values = np.min(A, axis=1).reshape(8, 1)

print("Матриця A (8x8):\n", A)
print("\nВектор-стовпчик з мінімальними елементами рядків:\n", min_values)
```
